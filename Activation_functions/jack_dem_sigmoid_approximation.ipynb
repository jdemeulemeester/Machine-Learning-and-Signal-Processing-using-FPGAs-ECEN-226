Least Squares Modeling of Sigmoid


```python
import numpy as np
from matplotlib import pyplot
import tensorflow as tf
```

```python
def sigmoid(x):
    return(1./(1. + np.exp(-x)))
```

```python
def tanh(x):
    return((np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))) # you need to modify this
```

```python
X = np.array([0, 2, 4])
#X[0]
A = np.array([[1, X[0], X[0]**2], [1, X[1], X[1]**2], [1, X[2], X[2]**2]])
b = np.array([sigmoid(X[0]), sigmoid(X[1]), sigmoid(X[2])])
c = np.linalg.solve(A, b)
print(A)
print(b)
print(c)
np.savetxt("sigmoid_coeffs.dat", c)
```

```python
N = 20
poly_apprx_max = 4
x_max = 8
x = np.linspace(-x_max, x_max, num=N)
sigmoid_apprx = np.zeros(N)

#c = c2

for n in range(N):
    m = np.abs(x[n])
    if m > poly_apprx_max:
        tmp = 1;
    else:
        tmp = c[0] + c[1]*m + c[2]*m**2

    if x[n] < 0.0:
        sigmoid_apprx[n] = 1.0-tmp
    else:
        sigmoid_apprx[n] = tmp

sigmoid_reference = sigmoid(x)
sigmoid_diff = sigmoid_reference - sigmoid_apprx

plot_nrows = 1
plot_ncols = 1

fig, axs = pyplot.subplots(2, 1)
axs[0].plot(x, sigmoid_apprx, x, sigmoid_reference)
#axs[0].set_xlim(0, 2)
axs[0].set_xlabel('x')
axs[0].set_ylabel('sigmoid(x)')
axs[0].grid(True)
axs[1].plot(x, sigmoid_diff)
axs[1].grid(True)
axs[1].set_xlabel('x')
axs[1].set_ylabel('difference')
```

```python
X = np.array([0, 0.4, 0.8])
#X[0]
A = np.array([[1, X[0], X[0]**2], [1, X[1], X[1]**2], [1, X[2], X[2]**2]])
b = np.array([tanh(X[0]), tanh(X[1]), tanh(X[2])])
c2 = np.linalg.solve(A, b)
print(A)
print(b)
print(c2)
np.savetxt("tanh_coeffs.dat", c2)
```

```python
N = 20
poly_apprx_max = 2
x_max = 8
x = np.linspace(-x_max, x_max, num=N)
tanh_apprx = np.zeros(N)

c = c2

for n in range(N):
    m = np.abs(x[n])
    if m > poly_apprx_max:
        tmp = 1;
    else:
        tmp = c[0] + c[1]*m + c[2]*m**2

    if x[n] < 0.0:
        tanh_apprx[n] =-tmp
    else:
        tanh_apprx[n] = tmp

tanh_reference = tanh(x)
tanh_diff = tanh_reference - tanh_apprx

plot_nrows = 1
plot_ncols = 1

fig, axs = pyplot.subplots(2, 1)
axs[0].plot(x, tanh_apprx, x, tanh_reference)
#axs[0].set_xlim(0, 2)
axs[0].set_xlabel('x')
axs[0].set_ylabel('tanh(x)')
axs[0].grid(True)
axs[1].plot(x, tanh_diff)
axs[1].grid(True)
axs[1].set_xlabel('x')
axs[1].set_ylabel('difference')
```
